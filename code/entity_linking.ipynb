```python
import spacy
from tqdm import tqdm
```

```python
# initialize language model
nlp = spacy.load("en_core_web_md")

# add pipeline (declared through entry_points in setup.py)
nlp.add_pipe("entityLinker", last=True)
```

```python
doc = nlp("I watched the Pirates of the Caribbean last silvester")
```

```python
with open('../data/wikidata5m_inductive/wikidata5m_text.txt') as text_in:

    page_links = {}

    for index, line in enumerate(tqdm(text_in, total=4815483)):
        # todo limit for debugging
        if index == 10:
            break

        uri, description = line.split('\t', 1)
        doc = nlp(description)


        page_links[uri] = []

        all_linked_entities = doc._.linkedEntities
        for entity in all_linked_entities:
            subj_uri = 'Q' + str(entity.get_id())
            print(subj_uri)
            page_links[uri].append(subj_uri)
        print('----')


```
